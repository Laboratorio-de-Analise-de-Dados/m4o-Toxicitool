# Model Results Visualization

This notebook visualizes the results from the trained models.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from pathlib import Path

# Set style for plots
plt.style.use('seaborn-v0_8')
sns.set_palette('husl')

# Define paths
project_root = Path(__file__).parent.parent
results_path = project_root / 'results'
print(f'Project root: {project_root}')
```

## Load Model Results

```python
# Try to load the model log
try:
    log_path = results_path / 'tabelas' / 'multitask' / 'MTL_Log_Modelos_v3.xlsx'
    if log_path.exists():
        df_results = pd.read_excel(log_path)
        print(f'Results loaded successfully. Shape: {df_results.shape}')
        print('\nColumns:')
        print(df_results.columns.tolist())
        print('\nFirst few rows:')
        print(df_results.head())
    else:
        print('Model results not found. Please run model training first.')
except Exception as e:
    print(f'Error loading results: {e}')
```

## Performance Visualization

```python
if 'df_results' in locals():
    # Get all R² and MAE columns
    r2_cols = [col for col in df_results.columns if col.startswith('r2_')]
    mae_cols = [col for col in df_results.columns if col.startswith('mae_')]
    
    print('R² columns:', r2_cols)
    print('MAE columns:', mae_cols)
    
    # Plot R² scores
    if r2_cols:
        plt.figure(figsize=(12, 6))
        df_results[r2_cols].boxplot()
        plt.title('Distribution of R² Scores by Model')
        plt.ylabel('R² Score')
        plt.xticks(rotation=45)
        plt.tight_layout()
        plt.show()
    
    # Plot MAE scores
    if mae_cols:
        plt.figure(figsize=(12, 6))
        df_results[mae_cols].boxplot()
        plt.title('Distribution of MAE Scores by Model')
        plt.ylabel('MAE Score')
        plt.xticks(rotation=45)
        plt.tight_layout()
        plt.show()
```

## Model Ranking

```python
if 'df_results' in locals():
    # Sort by R² score (descending)
    df_sorted = df_results.sort_values('r2_medio', ascending=False)
    
    print('Top 5 Models by R² Score:')
    print(df_sorted[['r2_medio', 'mae_medio', 'loss_global']].head())
    
    # Plot ranking
    plt.figure(figsize=(10, 6))
    plt.bar(range(len(df_sorted)), df_sorted['r2_medio'])
    plt.xlabel('Model Rank')
    plt.ylabel('R² Score')
    plt.title('R² Score by Model Rank')
    plt.tight_layout()
    plt.show()
```
